```python
directory='/content/drive/MyDrive/potato_disease_class'
dataset=tf.keras.preprocessing.image_dataset_from_directory(
    directory,
    labels='inferred',
    color_mode='rgb',
    batch_size=32,
    image_size=(256, 256),
    shuffle=True,
    seed=99,
)
```

```python
class_names=dataset.class_names
```

```python
train_size=0.8
test_size=0.1
val_size=0.1
```

```python
def partition_data(ds,train_size=0.8,test_size=0.1,val_size=0.1,shuffle=True,shuffle_size=1000):
  train_len=int(len(dataset)*train_size)
  val_len=int(len(dataset)*test_size)

  if shuffle:
    ds=ds.shuffle(shuffle_size,seed=10)

  train_ds=ds.take(train_len)
  rem_ds=ds.skip(train_len)
  val_ds=rem_ds.take(val_len)
  test_ds=rem_ds.skip(val_len)
  return train_ds,val_ds,test_ds
```

```python
train_ds,val_ds,test_ds=partition_data(dataset)
```

```python
len(val_ds)
```

```python
train_ds=(
    train_ds.cache().
    shuffle(1000).
    prefetch(tf.data.AUTOTUNE)
)
```

```python
train_ds
```

```python
val_ds=(
    val_ds.cache().
    shuffle(1000).
    prefetch(tf.data.AUTOTUNE)
)
```

```python
test_ds=(
    test_ds.cache().
    shuffle(1000).
    prefetch(tf.data.AUTOTUNE)
)
```

```python
Resize_Rescale=tf.keras.Sequential([
    Resizing(256,256),
    Rescaling(1./255.)
])
```

```python
augment_layers=tf.keras.Sequential([
    RandomRotation(factor=(0.2)),
    RandomFlip(mode='horizontal_and_vertical'),
])
```

```python
def augment_ds(image,label):
  return augment_layers(image, training=True),label
```

```python

from tensorflow.keras.layers import InputLayer, Conv2D, MaxPooling2D, BatchNormalization, Dense, Flatten
from tensorflow.keras.losses import CategoricalCrossentropy, SparseCategoricalCrossentropy
from tensorflow.keras.metrics import CategoricalAccuracy, TopKCategoricalAccuracy
from tensorflow.keras.optimizers import Adam
```

```python
model=tf.keras.Sequential([
    InputLayer(input_shape=(None,None,3),),
    Resize_Rescale,
    augment_layers,
    Conv2D(32,(3,3),activation='relu'),
    MaxPooling2D((2,2)),
    Conv2D(64,(3,3),activation='relu'),
    MaxPooling2D((2,2)),
    Conv2D(64,(3,3),activation='relu'),
    MaxPooling2D((2,2)),
    Conv2D(64,(3,3),activation='relu'),
    MaxPooling2D((2,2)),
    Conv2D(64,(3,3),activation='relu'),
    MaxPooling2D((2,2)),
    Conv2D(64,(3,3),activation='relu'),
    MaxPooling2D((2,2)),

    Flatten(),
    Dense(64,activation='relu'),
    Dense(3,activation='softmax')

])
```

```python
model.summary()
```

```python
loss_fxn=SparseCategoricalCrossentropy()
```

```python
metrics=[CategoricalAccuracy(name='accuracy'),TopKCategoricalAccuracy(k=2,name='topKaccuracy')]
```

```python
model.compile(
    optimizer='adam',
    loss=loss_fxn,
    metrics=['accuracy']
)
```

```python
history=model.fit(
    train_ds,
    validation_data=val_ds,
    epochs=50,
    verbose=1,
    batch_size=32
)
```

```python
model.save('/content/drive/MyDrive/potato_disease_class/model.h5')
```

```python
model.evaluate(test_ds)
```
